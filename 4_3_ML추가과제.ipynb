# 1. 데이터 준비

```python
import pandas as pd

iris = pd.read_csv('https://raw.githubusercontent.com/jmnote/zdata/master/R/iris.csv')
iris.head()
```

# 2. 클래스 추출


```python
print(iris.columns)
```

# 3. 입력 데이터와 정답 데이터를 생성

```python
iris_input = iris[['Sepal.Length',	'Sepal.Width',	'Petal.Length',	'Petal.Width']].to_numpy() # 입력 데이터 생성
iris_target = iris['Species'].to_numpy() # 타겟 데이터 생성
iris_target
```

# 4. 훈련 세트와 테스트 세트로 나누기

**훈련 데이터와 타겟 데이터로 데이터 분할하기**

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(iris_input, iris_target, random_state=42)
```

# 5. 데이터 표준화

**훈련 데이터와 타겟 데이터를 표준화 전처리 하기**

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler() # StandardScaler 클래스의 인스턴스를 생성합
ss.fit(train_input) # StandardScaler 인스턴스를 사용하여 학습 데이터인 train_input에 대한 평균과 표준편차를 계산
train_scaled = ss.transform(train_input) # 학습 데이터인 train_input을 스케일링하여 train_scaled에 저장
test_scaled = ss.transform(test_input) # 테스트 데이터인 test_input을 스케일링하여 test_scaled에 저장
```

##### scaled

```python
train_scaled
```

```python
test_scaled
```

# 6. 테스트 세트의 5개의 데이터에 대해서 예측과 확률을 알아보기

**K-NN 분류기로 타겟 값 예측하기**

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier(n_neighbors=3)
kn.fit(train_scaled, train_target)

print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))
```

```python
print(kn.classes_)
```

```python
print(kn.predict(test_scaled[:5]))
```

처음 5개 샘플의 타겟값 예측

**클래스 확률 알아보기**

```python
import numpy as np
proba = kn.predict_proba(test_scaled[:5])
print(np.round(proba, decimals=4))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터프레임에서 산점도 행렬 그리기
# 데이터프레임 df를 사용하고, vars는 그래프에 포함될 속성들의 리스트입니다.
sns.set(style="ticks")
g = sns.pairplot(data=iris, vars=['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width'], hue='Species')

# 첫 5개 데이터 포인트를 각각 다른 마커로 표시
markers = ['o', 's', 'D', '^', 'v']  # 다섯 가지 다른 마커 (원, 사각형, 다이아몬드, 삼각형, 역삼각형 등)
for i in range(5):
    # g.data과 같은 데이터프레임을 사용하여 첫 5개 데이터 포인트에 접근합니다.
    data_point = iris.iloc[i]

    # 원하는 모양으로 설정 (예: 삼각형 모양)
    g.axes[0, 0].scatter(data_point['Sepal.Length'], data_point['Sepal.Width'], marker=markers[i], s=50, color='red')
    g.axes[1, 1].scatter(data_point['Sepal.Width'], data_point['Petal.Length'], marker=markers[i], s=50, color='red')
    g.axes[2, 2].scatter(data_point['Petal.Length'], data_point['Petal.Width'], marker=markers[i], s=50, color='red')

# 범례 추가
plt.legend()

# 그래프 표시
plt.show()
```

# 7. 최적화하기

```python
iris_target.shape
```

```python
train_score = []
test_score = []

# 3부터 159의 제곱근 약 13까지
for i in range(3, 14):
  kn = KNeighborsClassifier(n_neighbors=i)
  kn.fit(train_scaled, train_target)
  print("---------------------")
  print(f"n_neighbors : {i}")
  train_score.append(kn.score(train_scaled, train_target))
  test_score.append(kn.score(test_scaled, test_target))
  print(kn.score(train_scaled, train_target))
  print(kn.score(test_scaled, test_target))
print("---------------------")
print(f" Best n_neighbors [train]: {train_score.index(max(train_score)) + 3}")
print(f" Best kn score [train]: {max(train_score)}")
if (max(test_score) == test_score[train_score.index(max(train_score))]):
  print(f" Best n_neighbors [test]: {train_score.index(max(train_score)) + 3}")
  print(f" Best kn score [test]: {max(test_score)}")
else:
  # 최대값을 역순으로 찾기 위해 리스트를 뒤집습니다.
  reversed_test_scores = test_score[::-1]
  # 최대값의 인덱스를 찾습니다.
  max_index = len(reversed_test_scores) - 1 - reversed_test_scores.index(max(reversed_test_scores))

  print(f" Best n_neighbors [test]: {max_index}")
  print(f" Best kn score [test]: {max(test_score)}")
```

# 8. 로지스틱 회귀 -> 다중 분류 훈련 및 최적화

**sigmoid 함수 그래프**

```python
import numpy as np
import matplotlib.pyplot as plt

z = np.arange(-5, 5, 0.1)
phi = 1 / (1 + np.exp(-z))

plt.plot(z, phi)
plt.xlabel('z')
plt.ylabel('phi')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(C=20, max_iter=1000)
lr.fit(train_scaled, train_target)

print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

```python
print(lr.predict(test_scaled[:5]))
```

```python
print(lr.classes_)
```

**Probability[예측] 배열**

```python
proba = lr.predict_proba(test_scaled[:5])
print(np.round(proba, decimals=3))
```

```python
print(lr.coef_.shape, lr.intercept_.shape)
```

```python
decision = lr.decision_function(test_scaled[:5])
print(np.round(decision, decimals=2))
```

softmax 함수를 이용해 구한 배열

```python
from scipy.special import softmax

proba = softmax(decision, axis=1)
print(np.round(proba, decimals=3))
```

-> 결과가 정확히 일치함을 확인할 수 있음

# 9. 확률적 경사 하강법 -> 다중 분류 훈련 및 최적화

```python
from sklearn.linear_model import SGDClassifier
```

```python
sc = SGDClassifier(loss='log_loss', max_iter=10, random_state=42)
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

```python
import numpy as np

sc = SGDClassifier(loss='log_loss', random_state=42)

train_score = []
test_score = []

classes = np.unique(train_target)
```

```python
for _ in range(0, 300):
    sc.partial_fit(train_scaled, train_target, classes=classes)

    train_score.append(sc.score(train_scaled, train_target))
    test_score.append(sc.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt

plt.plot(train_score)
plt.plot(test_score)
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

```python
import matplotlib.pyplot as plt

# train_score와 test_score 리스트가 이미 채워져 있다고 가정합니다.

# 에포크(반복) 횟수
epochs = range(1, len(train_score) + 1)

# 훈련 점수와 테스트 점수를 그래프로 표현
plt.plot(epochs, train_score, label='Train Score', marker='o')
plt.plot(epochs, test_score, label='Test Score', marker='x')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
